#ENCODING


## In encoding we need to convert the categorical data into numerica data or binary form.

```python
# In one hot encoding the columns get inceased so we go with label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['State']=le.fit_transform(data['State'])
data.head()
# Now the state categorical data convert into numeric data
```

##After encoding seperate the independent and dependent value

```python
y=data['Profit']
x=data.drop(columns=['Profit'],axis=1)
```

##Next we need to scale the data so the datA IS CONVERTED INTO ARRAY

```python
names=x.columns
names
```

##APPLYING SCALING FUNCTION

```python
from sklearn.preprocessing import scale
x=scale(x)
x
# Now the dataset is converted into an array . see the output.

```

```python
x=pd.DataFrame(x,columns=names)
x.head()
#Now again we convert it into dataframe from scale
```

Now we are going to create a model by using train test and spilt

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train
```

#ANN Creation

For deep learning model we have certain libraries

1.TensorFlow

2.Keras

```python
# We are going to use TensorFlow model
import tensorflow
# here we are using the sequential model like [input->hidden layers->output this kind is known as sequential mode]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
# In this model we need to create layers using Dense
```

#Creating the input layer

(here we are taking the four layers so units is 4)

```python
# Input layer - By using Dense
# Weights - By using kernel_initializer
# Activation function - relu (Positive means  positive if it is negative means Zero)

model.add(Dense(units = 4,kernel_initializer= "random_uniform",activation= "relu"))
```

# Creating the hidden layers 

(Here there is no restriction for the hidden layer so randomly we are taking the 10 hidden layers)

```python
# Hidden layer units is 10
# kernela and activation are same as input if we want we will change.

# First hidden layer
model.add(Dense(units = 10,kernel_initializer= "random_uniform",activation= "relu"))
```

```python
#  Second hidden layer
model.add(Dense(units = 12,kernel_initializer= "random_uniform",activation= "relu"))
```

#The third stage is output layer

here we are using the regression so there is no need for the multiple units only ine unit is enough

```python
# We dont need any activation madel in output layer
model.add(Dense(units = 1 ,kernel_initializer= "random_uniform"))
```

#Then we need to do compile the output by using the mean square error (mse)

```python
model.compile(loss="mse",metrics=["mse"],optimizer="adam")
```

After compile we need to train the data for that we need to fit it.

```python
model.fit(x_train,y_train,epochs=10,batch_size=10)
#epochs means one forward mevement and one backward movement is called as epochs.like it is used for the calculating error
```

Now we want to predict the values

```python
pred = model.predict(x_test)
pred
#this is our predicted value

# then we need to compare the original value and predicted value
```

#For comparision we need to use the inbuild module metrices

```python
from sklearn import metrics
metrics.mean_squared_error(pred,y_test)
# This is not a optimal model because there is more errors are there
```
