# We use Activation function to covert linear output(y=mx+=c) into non-linear output

for example: In water by adding the different incredient we make different juices like orange or lemon juice.

like that by adding activation function we make linear into non linear output(IN 0's and 1's)

## The name of the ACTIVATION FUNCTION IS relu activation function 

RELU-Rectified Linear Unit

this convert Linear function into non linear function

**From non linear output function we need to apply another activation function to get the desired output.**

### Output may be in the form of 
1.Continous - Salary

2.Binary - yes or no / 1 or 0

3.Categorical - name , place etc

**AI is also called as perceptron**

### **Types of activation function**

1.THERSHOLD FUNCTION  [0 and 1 binary]

2.RELU    [32 gives 32 , negatve value change into 0]

3.SIGMOID FUNCTION   [0 and 1 binary]

4.TANH FUNCTION  [-1,0,1]

5.SOFTMAX FUNCTION  [more then two categories like 1,2,3 or north east west south]
